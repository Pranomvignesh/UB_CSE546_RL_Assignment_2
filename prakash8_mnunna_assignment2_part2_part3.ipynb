```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import gym
import pickle
from dqn import Options,DQN,DuelingDQN,DoubleDQN,Hyperparams,EnvInfo
from gridEnvironment import GridEnvironment


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

## DQN Grid Environment

```python
class NeuralNetworkForGridEnv(nn.Module):
    def __init__(self, observation_space, action_space, learningRate):
        super().__init__()
        self.layer_1 = nn.Linear(observation_space, 256)
        self.layer_2 = nn.Linear(256, 256)
        self.layer_3 = nn.Linear(256, action_space)
        self.optimizer = optim.Adam(self.parameters(), lr=learningRate)
        self.loss = nn.MSELoss()
        self.to(device)

    def forward(self, x):
        x = F.relu(self.layer_1(x))
        x = F.relu(self.layer_2(x))
        return self.layer_3(x)

gridEnv = GridEnvironment(env="Deterministic", max_timesteps=100)
    
dqnGridEnv = DQN(
    envInfo=EnvInfo(
        env=gridEnv,
        observation_space = gridEnv.observation_space.n,
        action_space = gridEnv.action_space.n
    ),
    hyperparams=Hyperparams(
        epsilon=0.9,
        epsilonMin=0.001,
        epsilonDecay=1000,
        memorySize=1000,
        learningRate=1e-4,
        batchSize=128,
        discountFactor=0.99,
        targetNetworkUpdateFrequency=100,
        episodes=300
    ),
    nnModel=NeuralNetworkForGridEnv,
    options = Options(
        resultsPath='./results',
        filePrefix='GridEnv',
        showLiveResults=True,
        logResults = False,
        saveModels = True,
        debug = False
    )
)

dqnGridEnv.train()
```

## Saving DQN Grid Environment

```python
dqnGridEnv.saveWeights()
dqnGridEnv.saveModels()
dqnGridEnv.saveResults()
```

## Evaluation of DQN for Grid Environment

```python
policyModel = torch.load('./results/models/GridEnv_DQN_policy_model.pth')
targetModel = torch.load('./results/models/GridEnv_DQN_target_model.pth')

dqnGridEnv.loadModels(policyModel,targetModel)

dqnGridEnv.greedy(timeSteps = 15)
```

## DQN - Cartpole

```python
class NeuralNetworkForCartPole(nn.Module):
    def __init__(self, observation_space, action_space, learningRate):
        super().__init__()
        self.layer_1 = nn.Linear(observation_space, 128)
        self.layer_2 = nn.Linear(128, 128)
        self.layer_3 = nn.Linear(128, action_space)
        self.optimizer = optim.Adam(self.parameters(), lr=learningRate)
        self.loss = nn.MSELoss()
        self.to(device)

    def forward(self, x):
        x = F.relu(self.layer_1(x))
        x = F.relu(self.layer_2(x))
        return self.layer_3(x)
    
CartPoleParameters = (
    EnvInfo(env = gym.make('CartPole-v1')),
    Hyperparams(
        epsilon=0.99,
        epsilonMin=0.001,
        epsilonDecay=10000,
        memorySize=5000,
        learningRate=1e-4,
        batchSize=128,
        discountFactor=0.99,
        targetNetworkUpdateFrequency=30,
        episodes=700
    ),
    NeuralNetworkForCartPole,
    Options(
        resultsPath='./results',
        filePrefix='CartPole',
        showLiveResults=True,
        logResults = False,
        saveModels = True,
        debug = False
    )
)


dqnCartPole = DQN(*CartPoleParameters)

dqnCartPole.train()

```

```python
import matplotlib
import matplotlib.pyplot as plt
import pickle
from pathlib import Path

def compareRewards(environment,algorithms):
    resultsPath = './results'
    results = []
    for algorithm in algorithms:
        filePath = f'{resultsPath}/rewards/{environment}_{algorithm}_rewards.pkl'
        with open(filePath, 'rb') as f:
            _,averageRewards = pickle.load(f)
            results.append(averageRewards)
    
    
    minLength = min([len(x) for x in results])
    results = [ reward[-minLength:] for reward in results ]
    fig = plt.figure(1)
    plt.title(f'Comparison of Results for {environment}')
    plt.xlabel('Episodes')
    plt.ylabel('Average Rewards')
    fig.set_figwidth(8)
    fig.set_figheight(6)
    for reward in results:
        plt.plot(reward)
    plt.legend(algorithms)
    fileName = '_vs_'.join(algorithms)
    Path(f'{resultsPath}/images').mkdir(parents=True, exist_ok=True)
    plt.savefig(f'{resultsPath}/images/Comparison_for_{environment}_{fileName}.png')
    plt.show()
```

## Comparision between DQN and DoubleDQN in CartPole Environment

```python
compareRewards('CartPole',['DQN','DoubleDQN'])
```

## Comparision between DQN and DoubleDQN in Acrobot Environment

```python
compareRewards('Acrobot',['DQN','DoubleDQN'])
```

## Comparision between DQN and DoubleDQN in Grid Environment
